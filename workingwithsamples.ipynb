```python

import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.stem import PorterStemmer,WordNetLemmatizer
from nltk.corpus import stopwords
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
# Example text data
text = """Natural language processing (NLP) is a subfield of artificial intelligence (AI) 
that focuses on enabling computers to understand, interpret, and generate human language 
in a way that is both valuable and meaningful."""


# Tokenize using regexp tokenizer
tokenizer = RegexpTokenizer(r'\w+')
tokens = tokenizer.tokenize(text.lower())  # Convert to lowercase

# Remove stopwords
stop_words = set(stopwords.words('english'))
filtered_tokens = [token for token in tokens if token not in stop_words]

# Stemming using PorterStemmer
porter = PorterStemmer()
stemmed_tokens = [porter.stem(token) for token in filtered_tokens]

# Lemmatization using WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]

# Print results
print("Original Text:")
print(text)
print("\nTokenized Text:")
print(tokens)
print("\nFiltered Tokens (without stopwords):")
print(filtered_tokens)
print("\nStemmed Tokens:")
print(stemmed_tokens)
print("\nLemmatized Tokens:")
print(lemmatized_tokens)
```
